## TASK 4 - IMAGE TO PENCIL SKETCH WITH PYTHON
## LEVEL - BEGINNER

### IMPORTING THE LIBRARIES

```python
import cv2
from matplotlib import pyplot as plt
```

## READ THE IMAGE IN RGB FORMAT

```python
image = cv2.imread('MS-Dhoni- indian jersey.jpg')
plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB),cmap='gray')
plt.show()
```

## CREATING A NEW IMAGE BY CONVERTING THE ORIGINAL IMAGE TO GREYSCALE

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(10,10))
plt.imshow(gray_image,cmap='gray')
plt.show()
```

## INVERTING IMAGE OF MS DHONI

```python
inverted_image = 255 - gray_image
plt.figure(figsize=(10,10))
plt.imshow(inverted_image,cmap='gray')
plt.show()
```

## BLUR THE IMAGE OF MS DHONI BY USING THE GAUSSIAN FUNCTION

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
plt.figure(figsize=(10,10))
plt.imshow(blurred,cmap='gray')
plt.show()
```

## INVERT THE BLURRED IMAGE FOR FINAL IMAGE

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB),cmap='gray')
plt.show()
```

## FINAL IMAGE OF PENCIL SKETCH

```python
plt.figure(figsize=(10,10))
plt.imshow(pencil_sketch,cmap='gray')
plt.show()
```
